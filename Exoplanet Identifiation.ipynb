##### A/ The problem

For this final project, I have chosen to look at deep learning applications into exoplanets identification. The problem is simple: given star observation made by a Telescope (Kepler, TESS, etc...), can a deep learning model provide enough certainty as to identify exoplanets around it.

### 2/ Exploratory Data Analysis (EDA) â€” Inspect, Visualize and Clean the Data

##### A/ Import libraries

Let's start with the usual imports.

```python
# usual imports
import matplotlib as plt
import numpy as np
import pandas as pd
import random as rd
import seaborn as sns
import tensorflow as tf
import visualkeras

from sklearn.linear_model import LogisticRegression
from sklearn import metrics
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from tensorflow.keras.layers import Dense, Flatten, Dropout
from tensorflow.keras.models import Sequential
from tensorflow.keras.optimizers import Adam
```

Let's set up a seed for repeatable conclusion.

```python
# set seeds for repeatable conclusion
seed = 42
rd.seed(seed)
np.random.seed(seed)
tf.random.set_seed(seed)
```

##### B/ Import the dataset

Next, let's import the dataset.

```python
# import the dataset
exoplanet_df = pd.read_csv('exoplanet_dataset.csv')

# first look
exoplanet_df.head()
```

##### C/ Inspect the dataset

Let's have a look at the dataset. The goal is to identify:
* which column to use as result
* which columns are superfluous
* which columns have Null, NaN or blank cells requiring cleaning.

Before we go about cleaning the dataset, we can make the following observations:
- the result is the **koi_disposition** column. For simplicity's sake, we will reduce it to two values: Confirmed (=1), Negative (=0)
- all the identification columns must be removed, we need to do the classification based on interstellar measurement only

Let's removed the identification columns first.

```python
id_columns_to_remove = [
    'rowid',
    'kepid',
    'kepoi_name',
    'kepler_name',
    'koi_vet_stat',
    'koi_vet_date',
    'koi_pdisposition',
    'koi_score',
    'koi_fpflag_nt',
    'koi_fpflag_ss',
    'koi_fpflag_co',
    'koi_fpflag_ec',
    'koi_disp_prov',
    'koi_comment',
    'koi_limbdark_mod',
    'koi_trans_mod',
    'koi_datalink_dvr',
    'koi_datalink_dvs',
    'koi_sparprov',
    'koi_tce_delivname',
    'koi_fittype',
    'koi_parm_prov'
]

exoplanet_df.drop(id_columns_to_remove, axis=1, inplace=True)
```

Now we can clean the **koi_disposition** column.

```python
# find current values
exoplanet_df['koi_disposition'].unique()

# update dataframe column
exoplanet_df.replace('CONFIRMED', 1, inplace=True)
exoplanet_df.replace('CANDIDATE', 0, inplace=True)
exoplanet_df.replace('FALSE POSITIVE', 0, inplace=True)

# look at the distribution of the result
exoplanet_df['koi_disposition'].hist()
```

About a quarter of the data is identified as Exoplanets.

We can now shuffle and standardize the data.

```python
# shuffle the entire dataset manually
exoplanet_df = exoplanet_df.sample(frac=1).reset_index(drop=True)

# standardize the dataset
for column in exoplanet_df.columns:
    if column != 'koi_disposition':
        exoplanet_df[column] = (exoplanet_df[column] - exoplanet_df[column].mean()) / exoplanet_df[column].std()
```

Now, let's look at the numerical columns.

```python
# describe the statistic of the numerical column
exoplanet_df.describe()
```

We can see that 125 columns out of 140 are not numerical. We can also see that several columns do not have the maximum count of 9564.

Let's move onto cleaning.

##### D/ Cleaning the dataset

Let's get a sense of the amount of Null/NaN data.

```python
# display count and type per column
exoplanet_df.info(verbose=True, show_counts=True)
```

We can see four groups of counts:
* columns with full quantity (9564)
* columns with near full quantity (~8000)
* columns with near empty quantity (~2000)
* columns with empty quantity (0)

Let's start by removing the empty columns.

```python
# drop empty columns
exoplanet_df.dropna(how='all', axis=1, inplace=True)
```

Now, let's take a closer at the near full columns.

```python
# display non-null count per column
exoplanet_df[exoplanet_df.columns[exoplanet_df.isna().any()]].info()
```

We can see that the minimum is the column **koi_bin_oedp_sig** with 8054 non-null count. This represents (9564 - 8054) / 9564 * 100 = **15.8%**. This value is a bit too big, so let's select a treshold at 9000: (9564 - 9000) / 9564 * 100 = **5.9%**. We will update all those columns null cells with their respective mean.

```python
# prepare list of column to fill with respective mean
columns_to_fill = [
    'koi_max_sngle_ev',
    'koi_max_mult_ev',
    'koi_num_transits',
    'koi_quarters',
    'koi_bin_oedp_sig',
    'koi_zmag',
    'koi_fwm_stat_sig',
    'koi_fwm_prao',
    'koi_fwm_prao_err',
    'koi_fwm_pdeco',
    'koi_fwm_pdeco_err',
    'koi_dicco_mra',
    'koi_dicco_mra_err',
    'koi_dicco_mdec',
    'koi_dicco_mdec_err',
    'koi_dicco_msky',
    'koi_dicco_msky_err',
    'koi_dikco_mra',
    'koi_dikco_mra_err',
    'koi_dikco_mdec',
    'koi_dikco_mdec_err',
    'koi_dikco_msky',
    'koi_dikco_msky_err'
]

# fill columns with mean
for column in columns_to_fill:
    exoplanet_df[column] = exoplanet_df[column].fillna(exoplanet_df[column].mean())
```

The rest of the data that contains null values represent less than 10% of the overall dataset. It is now below the standard treshold and can be removed.

```python
# remove rows with NaN
exoplanet_df.dropna(axis=0, inplace=True)
```

We can observe that the _non-overlapping_ of those NaN value means we now have 8443 rows remaining. This represents (9564 - 8443) / 9564 * 100 = **11.7%**. While this value is above the standard 10%, this leaves enough data to analyse.

##### E/ Train/test split

With a clean dataset, we can now split our train/test sets.

```python
# retrieve result column
y_true = exoplanet_df['koi_disposition']

# drop the result column from the dataset
exoplanet_df.drop('koi_disposition', axis=1, inplace=True)

# split the train and test sets
x_train, x_test, y_train, y_test = train_test_split(exoplanet_df, y_true, shuffle=False, test_size=0.2, random_state=42)
```

Let's validate the split is successfull.

```python
# validation with shape
print(exoplanet_df.shape)
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

Finally, let's look at the distribution of the results.

```python
# distribution of y_true
y_true.hist()
```

The amound of positives has not changed, most of the cut out data where negative. The ratio is more balanced this way.

### 3/ Model Architecture

Given the data is non-visual and non temporal, a CNN or a RNN would be a poor fit. We will first run a standard logistic regression to have a base to compare with. Then, we will create a standard MLP, and tests several tuning of hyperparameters.

##### A/ Logistic Regression

We will start with a standard Logistic Regression in order to acquire a baseline to compare other algorithm to.

```python
# instance the model
model_logreg = LogisticRegression(solver='liblinear', random_state=42)

# train the model
model_logreg.fit(x_train, y_train)

# predict the test results
y_pred_logreg = model_logreg.predict(x_test)
```

Let's quickly evaluate the model.

```python
print("accuracy: " + str(metrics.accuracy_score(y_test, y_pred_logreg)))
print("confusion matrix:\n " + str(metrics.confusion_matrix(y_test, y_pred_logreg)))
```

**91.65%** is a fairly high ceiling to beat with so little code.

##### B/ Basic MLP

Let's see what would be the results of a simple MLP. We will use:
* 3 dense layers with relu activation
* 1 dense layer with sigmoid for classification

```python
# instance basic model
model_mlp1 = Sequential()
 
# add dense layers with relu activation
model_mlp1.add(Dense(100, activation='relu'))
model_mlp1.add(Dense(50, activation='relu'))
model_mlp1.add(Dense(10, activation='relu'))
 
# add a sigmoid layer for activation
model_mlp1.add(Dense(1, activation='sigmoid'))

# build the model
model_mlp1.build(input_shape=x_train.shape)
 
# model summary
model_mlp1.summary()
```

Let's get a better look at the model.

```python
# model visualization
visualkeras.layered_view(model_mlp1)
```

Now, let's choose some hyperparameters:
* binary classification lends itself well to binary_crossentropy
* adam is a good default
* we want to see the accuracy evolution through the epochs

```python
# compile the model
model_mlp1.compile(
    loss='binary_crossentropy',
    optimizer='adam',
    metrics=['accuracy'])
```

Let's fit our model with 10 epochs. Since the data has a small memory footprint, we can limit the batch size to 1.

```python
# train the model
history1 = model_mlp1.fit(x_train, y_train, epochs=10, batch_size=1, verbose=1, validation_data=(x_test, y_test))
```

Now let's display some simple results.

```python
# retrieve and display predictions
y_pred_mlp1 = [0 if x < .5 else 1 for x in model_mlp1.predict(x_test)]
print("accuracy: " + str(metrics.accuracy_score(y_test, y_pred_mlp1)))
print("confusion matrix:\n " + str(metrics.confusion_matrix(y_test, y_pred_mlp1)))
```

From those early results, we can make the following observation:
* our simple MLP reached an accuracy of nearly **93%**, and therefore does better than the logistic regression
* the confusion matrix is less balanced than the logistic regression model
* the accuracy of the train set is very close to the test set, and there is probably little to no overfitting. We can add a _dropout_ layer and see if we can reduce it further
* the accuracy for both the train set and the test set went up then down. It is not clear if the MLP has reached its limit, or if additional epochs could further increase the accuracy. We could also make the learning rate of adam smaller

##### C/ Better MLP

Based on the previous model, we will create a second model with the following modifications:
* addition of a _dropout_ layer
* reduction of the learning rate of the optimizer
* addition of 5 epochs
* addition of two _dense_ layers

```python
# instance basic model
model_mlp2 = Sequential()
 
# add dense layers with relu activation
model_mlp2.add(Dense(150, activation='relu'))
model_mlp2.add(Dense(120, activation='relu'))
model_mlp2.add(Dense(80, activation='relu'))
model_mlp2.add(Dense(50, activation='relu'))
model_mlp2.add(Dense(10, activation='relu'))

# add a dropout layer to reduce overfitting
model_mlp2.add(Dropout(0.3))
 
# add a sigmoid layer for activation
model_mlp2.add(Dense(1, activation='sigmoid'))

# build the model
model_mlp2.build(input_shape=x_train.shape)
 
# model summary
model_mlp2.summary()
```

Let's get a better look at the model.

```python
# model visualization
visualkeras.layered_view(model_mlp2)
```

Now, let's tune our hyperparameters.

```python
# let's customize our optimizer
optimizer = Adam(learning_rate=0.0001)

# compile the model
model_mlp2.compile(
    loss='binary_crossentropy',
    optimizer=optimizer,
    metrics=['accuracy'])
```

Let's add 5 epochs and train the model.

```python
# train the model
history2 = model_mlp2.fit(x_train, y_train, epochs=15, batch_size=1, verbose=1, validation_data=(x_test, y_test))
```

Now let's display some simple results.

```python
# retrieve and display predictions
y_pred_mlp2 = [0 if x < .5 else 1 for x in model_mlp2.predict(x_test)]
print("accuracy: " + str(metrics.accuracy_score(y_test, y_pred_mlp2)))
print("confusion matrix:\n " + str(metrics.confusion_matrix(y_test, y_pred_mlp2)))
```

The additional layers allows the train accuracy to get higher and faster than in the previous model, clearly allowing the model to learn more about our data. However, we can see that this accuracy on the test set was reached in the first batch of epochs, meaning the addition of 5 epochs was not needed. The difference between train and test set accuracy is similar to previously, so the _dropout_ layer doesn't seem to have much effect. Finally, the false positives and false negative are better balanced. In conclusion, this is a better, most robust model, but not by much.

### 4/ Results and Analysis

Let's put together our results.

```python
# plot results
logreg_acc = metrics.accuracy_score(y_test, y_pred_logreg)
plt.pyplot.plot([0, 15], [logreg_acc, logreg_acc], linestyle='-', linewidth=1)
plt.pyplot.plot(history1.history['accuracy'])
plt.pyplot.plot(history1.history['val_accuracy'])
plt.pyplot.plot(history2.history['accuracy'])
plt.pyplot.plot(history2.history['val_accuracy'])
plt.pyplot.title('models accuracy')
plt.pyplot.ylabel('accuracy')
plt.pyplot.xlabel('epoch')
plt.pyplot.legend(['logistic regression', 'train1', 'val1', 'train2', 'val2'], loc='lower right')
plt.pyplot.show()
```

We can see that both MLP outperform the Logistic Regression model, but the performance between both are very slim. It is also unclear which of the two MLP model provide less overfitting.

### 5/ Conclusion

We have seen that a simple Logistic Regression model provide excellent performance without spending much time on the tuning of hyperparameters. However, a simple MLP performs better than the Logistic Regression model with a very simple structure. Our second attempt to improve the MLP model provide little to no gains compare to the first one, meaning it is possible the data -a simple numeric table, however big- provides little complex structure to rely on for improvement through hyperparameters tuning. We can, however, make the model more robust against overfitting.

In conclusion, an MLP model is a good fit to the problem of classifying exoplanet based on stellar data. It is however quickly reaching its limits, and we suspect that domain specific knowledge would be required to better preprocess the data, in order to gain higher accuracy score.
